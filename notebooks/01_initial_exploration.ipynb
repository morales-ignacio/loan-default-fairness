```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 200)
pd.set_option('display.max_rows', 100)
sns.set_style('whitegrid')

print("Setup complete")
```

```python
df = pd.read_csv(
    'data/accepted_2007_to_2018Q4.csv',
    low_memory=False
)

print(f"Loaded {len(df):,} rows and {len(df.columns)} columns")
```

```python
# Filter to loans with final outcomes only
# "Current" loans haven't finished yet — we don't know if they'll default
completed_statuses = ['Fully Paid', 'Charged Off']
df_completed = df[df['loan_status'].isin(completed_statuses)].copy()

# Create binary target: 1 = defaulted, 0 = paid
df_completed['default'] = (df_completed['loan_status'] == 'Charged Off').astype(int)

# Report what we have
print(f"Original dataset:        {len(df):,} rows")
print(f"After filtering:         {len(df_completed):,} rows")
print(f"Removed (not completed): {len(df) - len(df_completed):,} rows")
print(f"\nOverall default rate:    {df_completed['default'].mean():.2%}")
```

```python
# Extract year from the issue_d (issue date) column
df_completed['issue_year'] = pd.to_datetime(df_completed['issue_d'], format='%b-%Y').dt.year

# Default rate by year, with loan counts
year_stats = df_completed.groupby('issue_year').agg(
    n_loans=('default', 'size'),
    default_rate=('default', 'mean')
).round(4)

print(year_stats)
```

```python
grade_stats = df_completed.groupby('grade').agg(
    n_loans=('default', 'size'),
    default_rate=('default', 'mean')
).round(4)

print(grade_stats)
```
